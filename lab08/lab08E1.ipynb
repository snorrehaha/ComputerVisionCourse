```python

from statistics import linear_regression

#Lab 08
import numpy as np
import matplotlib.pyplot as plt
import cv2
from sklearn.linear_model import LinearRegression
```

```python
#Create Dataset
np.random.seed(10)
x = np.random.rand(50,1) * 10
noise = np.random.randn(50,1) * 2
y = 3 * x + noise

plt.scatter(x,y, label="Data points") # Place dataset

model = LinearRegression() #Linear regression and fitting
model.fit(x,y)

a = model.coef_[0][0]
b = model.intercept_[0]

print(f"Slope (a): {a:.2f}")    #Store results
print(f"Intercept (b): {b:.2f}")

# Plot regression line
x_line = np.linspace(0, 10, 100).reshape(-1, 1)
y_line = model.predict(x_line)

plt.plot(x_line, y_line, color='red', label="Regression line")

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Linear Regression Example")
plt.show()
```

```python
y_pred = model.predict(x)

#SSE
sse = np.sum((y-y_pred) **2)
print(f"SSE: {sse:.2f}")

n = len(y)

mse = sse / n
rmse = np.sqrt(mse)

print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
```

```python
# Stronger noise
noise_strong = np.random.randn(50, 1) * 6
y_noisy = 3 * x + noise_strong

plt.scatter(x,y_noisy, label="Data points")

# Refit
model2 = LinearRegression()
model2.fit(x, y_noisy)

# Predictions
y_pred2 = model2.predict(x)

# Errors
sse2 = np.sum((y_noisy - y_pred2) ** 2)
mse2 = sse2 / len(y_noisy)
rmse2 = np.sqrt(mse2)

print("\nWith stronger noise:")
print(f"SSE: {sse2:.2f}")
print(f"MSE: {mse2:.2f}")
print(f"RMSE: {rmse2:.2f}")

x_line = np.linspace(0, 10, 100).reshape(-1, 1)
y_line = model2.predict(x_line)

#plotting
plt.plot(x_line, y_line, color='red', label="Regression line")

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Linear Regression Example")
plt.show()
```

8. What is the geometric interpretation of the parameters a and b in the model y = ax + b?
    - slope a:
        - a represents the rate of change of y with respect to x
        - a > 0: line go upwards
        - a < 0: Line go downwards
        - Larger a gives a steeper line
    - intercept b:
        - Where the line crosses the y-axis
        - The value of y when x = 0
        - it changes the line up or down without changing its slope
    - a controls the direction and steepness, b controls vertical position
9. Why are squared errors commonly used instead of absolute errors?
    - Squared error is smooth and differantiable everywhere
    - Absolute value is not differntiable at 0
    - With squared errors you get an formula for a,b
    - with absolute errors, its required iterative optimizaiton
    - Squared errors amplifies big deviations, forcing the model to avoid large mistakes
    - Squared errors is sensitive to outliers
