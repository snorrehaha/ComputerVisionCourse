```python
from statistics import linear_regression

#Lab 08
import numpy as np
import matplotlib.pyplot as plt
import cv2
from sklearn.linear_model import LinearRegression



from sklearn.datasets import make_blobs
from sklearn.datasets import make_classification


```

```python
np.random.seed(10)
X, y = make_blobs(n_samples=100, n_features=2, centers=2, cluster_std=1.5)

plt.scatter(X[y==0, 0], X[y==0, 1], color='red', label='Class 0')
plt.scatter(X[y==1, 0], X[y==1, 1], color='blue', label='Class 1')
plt.title('Toy 2D Dataset')
plt.legend()
plt.show()
```

```python
# Logistic regression
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(X,y)

print(f"Coefficients: {clf.coef_}, Intercept: {clf.intercept_}")
```

```python
#Linear SVM
from sklearn.svm import SVC
clf = SVC(kernel='linear')
clf.fit(X, y)

print(f"Coefficients: {clf.coef_}, Intercept: {clf.intercept_}")
```

```python
# Create meshgrid for plotting
x_min, x_max = X[:,0].min() - 1, X[:,0].max() + 1
y_min, y_max = X[:,1].min() - 1, X[:,1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200),
                     np.linspace(y_min, y_max, 200))

# Predict class on each point in the grid
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundary
plt.contourf(xx, yy, Z, alpha=0.2, colors=['red','blue'])
plt.scatter(X[y==0, 0], X[y==0, 1], color='red', label='Class 0')
plt.scatter(X[y==1, 0], X[y==1, 1], color='blue', label='Class 1')
plt.title('Linear Classifier Decision Boundary')
plt.legend()
plt.show()
```

```python
# Non-linearly separable dataset
from sklearn.datasets import make_circles
X_nl, y_nl = make_circles(n_samples=100, noise=0.1, factor=0.5, random_state=42)

plt.scatter(X_nl[y_nl==0,0], X_nl[y_nl==0,1], color='red', label='Class 0')
plt.scatter(X_nl[y_nl==1,0], X_nl[y_nl==1,1], color='blue', label='Class 1')
plt.title('Non-linearly Separable Dataset')
plt.legend()
plt.show()
```

```python
# Train a linear classifier (logistic regression or linear SVM)
clf_nl = LogisticRegression()
clf_nl.fit(X_nl, y_nl)

# Decision boundary
x_min, x_max = X_nl[:,0].min() - 1, X_nl[:,0].max() + 1
y_min, y_max = X_nl[:,1].min() - 1, X_nl[:,1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200),
                     np.linspace(y_min, y_max, 200))
Z = clf_nl.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot
plt.contourf(xx, yy, Z, alpha=0.2, colors=['red','blue'])
plt.scatter(X_nl[y_nl==0,0], X_nl[y_nl==0,1], color='red', label='Class 0')
plt.scatter(X_nl[y_nl==1,0], X_nl[y_nl==1,1], color='blue', label='Class 1')
plt.title('Linear Classifier on Non-linearly Separable Data')
plt.legend()
plt.show()
```
