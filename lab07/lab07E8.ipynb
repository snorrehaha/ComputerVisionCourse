```python
#Lab 07
import numpy as np
import matplotlib.pyplot as plt
import cv2
from skimage.feature import hog
from skimage import exposure


# FUNCTIONS ----------------------------------------------------------------------------------------
# Both LoadImage and GetSobelFeatures are from assignment 1: modified a bit
def LoadImage(path, size=(256, 256)):
    img = cv2.imread(path)
    img = cv2.resize(img, size)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    return img


def plot_images_grid(images, titles, cols=4, figsize=(12,6), heatmap=False):
    rows = int(np.ceil(len(images) / cols))
    plt.figure(figsize=figsize)

    for i, (img, title) in enumerate(zip(images, titles)):
        plt.subplot(rows, cols, i+1)
        plt.imshow(img, cmap='viridis' if heatmap else 'gray')
        plt.title(title)
        plt.axis("off")
        if heatmap:
            plt.colorbar()
    plt.tight_layout()
    plt.show()

def otsu_threshold(img):
    best_threshold = 0
    best_variance = float('inf')

    pixels = img.ravel()
    total_pixels = len(pixels)

    for t in range(0,256):
        class0 = pixels[pixels < t]
        class1 = pixels[pixels >= t]

        w0 = len(class0) / total_pixels
        w1 = len(class1) / total_pixels

        var0 = np.var(class0) if len(class0) > 0 else 0
        var1 = np.var(class1) if len(class1) > 0 else 0

        intra_class_variance = w0 * var0 + w1 * var1

        if intra_class_variance < best_variance:
            best_variance = intra_class_variance
            best_threshold = t
    return best_threshold
```

```python

img = LoadImage('4_coins.jpg')

gt = np.zeros_like(img)
gt[img > 120] = 255

_, otsu_mask = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Reshape image to 1D
Z = img.reshape((-1,1)).astype(np.float32)
K = 2
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
_, labels, centers = cv2.kmeans(Z, K, None, criteria, 10, cv2.KMEANS_PP_CENTERS)
# Convert back to image
labels = labels.reshape(img.shape)
# Convert to binary mask (choose one cluster as foreground)
kmeans_mask = np.uint8(labels == 1) * 255

def compute_metrics(gt, pred):
    gt = gt > 0
    pred = pred > 0

    TP = np.logical_and(gt, pred).sum()
    FP = np.logical_and(~gt, pred).sum()
    FN = np.logical_and(gt, ~pred).sum()
    TN = np.logical_and(~gt, ~pred).sum()

    accuracy = (TP + TN) / (TP + TN + FP + FN)
    iou = TP / (TP + FP + FN + 1e-8)
    dice = (2 * TP) / (2 * TP + FP + FN + 1e-8)

    return accuracy, iou, dice, TP, FP, FN, TN

metrics_otsu = compute_metrics(gt, otsu_mask)
metrics_kmeans = compute_metrics(gt, kmeans_mask)

print("Otsu:", metrics_otsu[:3])
print("K-means:", metrics_kmeans[:3])

def visualize_errors(img, gt, pred):
    gt = gt > 0
    pred = pred > 0

    TP = np.logical_and(gt, pred)
    FP = np.logical_and(~gt, pred)
    FN = np.logical_and(gt, ~pred)

    overlay = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)

    overlay[TP] = [0,255,0]   # Green
    overlay[FP] = [255,0,0]   # Red
    overlay[FN] = [0,0,255]   # Blue

    return overlay

vis_otsu = visualize_errors(img, gt, otsu_mask)
vis_kmeans = visualize_errors(img, gt, kmeans_mask)

plot_images_grid(
    [vis_otsu, vis_kmeans],
    ["Otsu Errors", "K-means Errors"],
    cols=2
)

def background_accuracy(gt, pred):
    gt = gt > 0
    pred = pred > 0

    TN = np.logical_and(~gt, ~pred).sum()
    total_bg = (~gt).sum()

    return TN / total_bg

print("Otsu BG accuracy:", background_accuracy(gt, otsu_mask))
print("K-means BG accuracy:", background_accuracy(gt, kmeans_mask))
```

1. Can a segmentation achieve high pixel accuracy but low IoU?

Yes. This occurs in class-imbalanced images where the background dominates.

Example:
If 95% of pixels are background and the model predicts all pixels as background:

Accuracy = 95% (high)
IoU (foreground) = 0 (very low)

The model ignores the object but still achieves high accuracy.

2. When can IoU and Dice give different rankings?

IoU and Dice differ in how they penalize errors:

IoU penalizes FP and FN more strictly
Dice is more forgiving due to the factor 2 * TP

They may rank methods differently when:
- One method has balanced FP and FN
- Another has high TP but also higher FP

Dice may favor the second (higher TP), while IoU penalizes it more.

3. Limitations of pixel-level metrics & additional criteria
Limitations:
Ignore spatial structure (shape, boundaries)
Do not reflect object-level correctness
Sensitive to class imbalance
Treat all pixels equally (boundary errors ≈ interior errors)
